彩色图像的灰度化、二值化


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from skimage.color import rgb2gray
```

```python
# Read image
img = Image.open('lenna.png')

# 将 PIL 图像对象转换为 NumPy 数组
img_array = np.array(img)

# 显示原始图像
plt.imshow(img_array)
plt.title('Original Image')
plt.show()
```

```python
# 灰度化
# Convert the image to grayscale (scikit-image)
img_gray_1 = rgb2gray(img_array)

# Convert the image to grayscale (OpenCV)
img_gray_2 = cv2.cvtColor(img_array, cv2.COLOR_BGR2GRAY)

# Convert the image to grayscale (NumPy)
img_gray_3 = np.dot(img_array[...,:3], [0.2989, 0.5870, 0.1140])


plt.figure(figsize=(15, 5))

# 显示灰度图像 #1
plt.subplot(1, 3, 1)
plt.imshow(img_gray_1, cmap='gray')
plt.title('Grayscale #1 scikit-image')

# 显示灰度图像 #2
plt.subplot(1, 3, 2)
plt.imshow(img_gray_2, cmap='gray')
plt.title('Grayscale #2 OpenCV')

# 显示灰度图像 #3
plt.subplot(1, 3, 3)
plt.imshow(img_gray_3, cmap='gray')
plt.title('Grayscale #3 NumPy')



# 显示图像
plt.show()
```

```python
# 打印图像的数据
print('img_gray_1:')
print(img_gray_1)

print('img_gray_2:')
print(img_gray_2)

print('img_gray_3:')
print(img_gray_3)
```

```python
# 二值化
# 应用阈值进行二值化 #1
# 参数：源图像，阈值，最大值，阈值类型
_, img_binary_1 = cv2.threshold(img_gray_2, 127, 255, cv2.THRESH_BINARY)

#打印二值化后的数据
print('img_binary_1:')
print(img_binary_1)



# 应用阈值进行二值化 #2
img_binary_2 = np.where(img_gray_2 > np.mean(img_gray_2), 1, 0)

# 打印二值化后的数据
print('img_binary_2:')
print(img_binary_2)



# 应用阈值进行二值化 #3
img_binary_3 = np.where(img_gray_1 >= 0.5, 1, 0)

# 打印二值化后的数据
print('img_binary_3:')
print(img_binary_3)



# 显示结果
plt.figure(figsize=(15, 5))

# 显示二值化图像 #1
plt.subplot(1, 3, 1)
plt.imshow(img_binary_1, cmap='gray')
plt.title('Binary Image #1')

# 显示二值化图像 #2
plt.subplot(1, 3, 2)
plt.imshow(img_binary_2, cmap='gray')
plt.title('Binary Image #2')

# 显示二值化图像 #3
plt.subplot(1, 3, 3)
plt.imshow(img_binary_3, cmap='gray')
plt.title('Binary Image #3')

# 显示图像
plt.show()
```
